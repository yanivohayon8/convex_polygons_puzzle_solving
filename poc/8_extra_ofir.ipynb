```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def match_piece_to_img(piece, img="", x_offset=0, y_offset=0, scale_factor=1.0, should_denormalize=True, show_img=False):
    new_piece = np.copy(piece)
    if scale_factor != 1:
        for i, vertice in enumerate(piece):
            # new_piece[i] = vertice * scale_factor
            new_piece[i][0] = vertice[0] * scale_factor
            new_piece[i][1] = vertice[1] * scale_factor
    if should_denormalize:
        denormalization_value = np.min(new_piece, axis=0)
    else:
        denormalization_value = np.array([0, 0])
    new_piece[:, 0] = new_piece[:, 0] - denormalization_value[0] + x_offset
    new_piece[:, 1] = new_piece[:, 1] - denormalization_value[1] + y_offset
    if show_img:
        plt.imshow(img)
        plt.fill(
                new_piece[:, 0],
                new_piece[:, 1],
                alpha=0.5,
                color='orange')
        plt.show()
    return new_piece
```

```python
# The coordinates from the original downloaded files
piece_0 = [(-187.0953374,-922.9101062),
(-466.3569015,-522.492106),
(-140.792368,1039.157992),
(794.244607,406.2442201)]

x_offset = 108
y_offset = 26
scale_factor = 0.2343980343980344
img_path = "C:\\Users\\97254\\Desktop\\msc\\Thesis\\Extrapolations\\puzzle_19_noise_0\\0_ext.png"
img = cv2.imread(img_path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
match_piece_to_img(piece_0,img,x_offset=x_offset,y_offset=y_offset,scale_factor=scale_factor,show_img=True)
```
